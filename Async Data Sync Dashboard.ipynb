```python
import asyncio
import random
from dotenv import load_dotenv
import os
from datetime import datetime
```

```python
load_dotenv()  # load from .env

API_KEY_1 = os.getenv("API_KEY_1")
API_KEY_2 = os.getenv("API_KEY_2")
API_KEY_3 = os.getenv("API_KEY_3")

print("Credentials loaded successfully")
```

```python
async def fetch_source1():
    await asyncio.sleep(random.uniform(0.5, 1.5))  # simulate network delay
    return {"source": "S1", "data": random.randint(10, 100)}

async def fetch_source2():
    await asyncio.sleep(random.uniform(0.5, 1.5))
    return {"source": "S2", "data": random.randint(100, 200)}

async def fetch_source3():
    await asyncio.sleep(random.uniform(0.5, 1.5))
    return {"source": "S3", "data": random.randint(200, 300)}
```

```python
def merge_data(results):
    merged = {}
    total = 0
    for r in results:
        merged[r['source']] = r['data']
        total += r['data']
    merged['total'] = total
    return merged
```

```python
async def run_dashboard():
    # Run all sources concurrently
    results = await asyncio.gather(
        fetch_source1(),
        fetch_source2(),
        fetch_source3()
    )
    
    merged = merge_data(results)
    timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    
    print(f"\n=== Async Data Sync Dashboard ({timestamp}) ===")
    for k, v in merged.items():
        print(f"{k}: {v}")
```

```python
asyncio.run(run_dashboard())
```
