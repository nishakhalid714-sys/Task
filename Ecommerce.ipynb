```python
# Sample messy product list
products = [
    # Electronics
    {"name": "Laptop", "category": "Electronics", "price": 1200},
    {"name": "Laptop", "category": "Electronics", "price": 1200},  # duplicate
    {"name": "Phone", "category": "Electronics", "price": 800},
    {"name": "Headphones", "category": "Electronics", "price": 150},
    {"name": "Monitor", "category": "Electronics", "price": 300},
    {"name": "Keyboard", "category": "Electronics", "price": None},  # missing price
    {"name": "Mouse", "category": "Electronics", "price": 50},
    {"name": "Webcam", "category": "Electronics", "price": 100},
    {"name": "Speaker", "category": "Electronics", "price": 200},
    # Home
    {"name": "Sofa", "category": "Home", "price": 700},
    {"name": "Sofa", "category": "Home", "price": 700},  # duplicate
    {"name": "Blender", "category": "Home", "price": 120},
    {"name": "Microwave", "category": "Home", "price": 250},
    {"name": "Lamp", "category": "Home", "price": None},  # missing price
    {"name": "Table", "category": "Home", "price": 400},
    {"name": "Chair", "category": "Home", "price": 150},
    {"name": "Vacuum", "category": "Home", "price": 300},
    # Books
    {"name": "Book A", "category": "Books", "price": 20},
    {"name": "Book B", "category": "Books", "price": 15},
    {"name": "Book C", "category": "Books", "price": None},  # missing price
    {"name": "Book D", "category": "Books", "price": 25},
    {"name": "Book E", "category": "Books", "price": 30},
    {"name": "Book F", "category": "Books", "price": 18},
    {"name": "Book G", "category": "Books", "price": 22},
    {"name": "Book H", "category": "Books", "price": 28},
]
```

 Clean data (remove duplicates and items with missing price)
 

```python
unique_products = { (p['name'], p['category']): p for p in products if p['price'] is not None }.values()
```

 Group by category
 

```python
from collections import defaultdict

grouped = defaultdict(list)
for p in unique_products:
    grouped[p['category']].append(p)
```

Sort each category by price descending and pick top 5

```python
summary = {cat: sorted(items, key=lambda x: x['price'], reverse=True)[:5] for cat, items in grouped.items()}
```

```python
for cat, items in summary.items():
    print(f"Category: {cat}")
    for item in items:
        print(f"  {item['name']} - ${item['price']}")
```
