## Task 2: Student Result Processor

```python
import os
import csv
```

```python
folder = r"C:\Users\nisha\Desktop\StudentData"
```

```python
output_file = "result.csv"

out = open(output_file, "w", newline="")
writer = csv.writer(out)

writer.writerow(["Name","Marks","Status"])
```

```python
for file in os.listdir(folder):
    if file.endswith(".csv"):
        path = os.path.join(folder,file)

        with open(path,"r") as f:
            reader = csv.reader(f)
            next(reader)

            for row in reader:
                name = row[0]
                marks = int(row[1])

                if marks >= 50:
                    status = "Pass"
                else:
                    status = "Fail"

                writer.writerow([name,marks,status])
```

```python
out.close()
print("Task Completed")
```
