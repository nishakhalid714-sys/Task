 Project Directory Auditor

```python
import os
from datetime import datetime, timedelta

folder = "."   # current folder
report_file = "audit_report.txt"

with open(report_file, "w") as report:
    
    for file in os.listdir(folder):
        
        if os.path.isfile(file):
            
            stats = os.stat(file)
            
            size = stats.st_size
            
            modified_time = datetime.fromtimestamp(stats.st_mtime)
            
            report.write(f"File: {file}\n")
            report.write(f"Size: {size} bytes\n")
            report.write(f"Last Modified: {modified_time}\n")
            
            # check if file older than 30 days
            if datetime.now() - modified_time > timedelta(days=30):
                report.write("⚠ OLD FILE (More than 30 days)\n")
            
            report.write("\n")

print("Audit Report Created Successfully ✅")
```
