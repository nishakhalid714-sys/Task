Sample Raw Orders

```python
orders = [
    {"customer": "Alice", "category": "Electronics", "price": 1200},
    {"customer": "Bob", "category": "Home", "price": 700},
    {"customer": "Charlie", "category": "Books", "price": 40},
    {"customer": "David", "category": "Electronics", "price": 800},
    {"customer": "Eva", "category": "Books", "price": 25},
    {"customer": "Frank", "category": "Home", "price": 150},
    {"customer": "Grace", "category": "Electronics", "price": 1500},
]
```

Discount Rules

```python
discount_rules = {
    "Electronics": 0.10,  # 10% discount
    "Home": 0.05,         # 5% discount
    "Books": 0.20         # 20% discount
}
```

Discounts (using map)

```python
from functools import reduce

# Apply discount
discounted_orders = list(map(
    lambda o: {**o, "discounted_price": o["price"] * (1 - discount_rules[o["category"]])},
    orders
))
```

Filter Orders above Minimum Threshold

```python
min_threshold = 100

valid_orders = list(filter(
    lambda o: o["discounted_price"] >= min_threshold,
    discounted_orders
))
```

Sort Orders by Discounted Price Descending

```python
sorted_orders = sorted(valid_orders, key=lambda o: o["discounted_price"], reverse=True)
```

Total Revenue (using reduce)

```python
total_revenue = reduce(lambda x,y: x + y["discounted_price"], sorted_orders, 0)
```

```python
print("Processed Orders:")
for o in sorted_orders:
    print(f"{o['customer']} - {o['category']} - Original: ${o['price']} - Discounted: ${o['discounted_price']:.2f}")

print(f"\nTotal Revenue: ${total_revenue:.2f}")
```
