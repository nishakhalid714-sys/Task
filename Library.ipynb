```python
from datetime import datetime, timedelta
```

```python
# Base Book class
class Book:
    def __init__(self, title, author, copies=1):
        self.title = title
        self.author = author
        self.copies = copies
    
    def __str__(self):
        return f"Book: {self.title} by {self.author} | Copies: {self.copies}"
    
    def __repr__(self):
        return f"<Book {self.title} ({self.copies} copies)>"
```

base member class

```python
class Member:
    def __init__(self, name, member_id):
        self.name = name
        self.member_id = member_id
        self.borrowed_books = {}  # {book: due_date}
    
    def borrow_book(self, book, days=14):
        if book.copies > 0:
            if len(self.borrowed_books) >= self.max_books():
                print(f"{self.name} has reached borrow limit!")
                return
            due_date = datetime.now() + timedelta(days=days)
            self.borrowed_books[book] = due_date
            book.copies -= 1
            print(f"{self.name} borrowed {book.title} | Due: {due_date.date()}")
        else:
            print(f"{book.title} is not available")
    
    def return_book(self, book):
        if book in self.borrowed_books:
            due_date = self.borrowed_books.pop(book)
            book.copies += 1
            fine = self.calculate_fine(due_date)
            if fine > 0:
                print(f"{self.name} returned {book.title} late | Fine: ${fine}")
            else:
                print(f"{self.name} returned {book.title} on time")
        else:
            print(f"{self.name} did not borrow {book.title}")
    
    def calculate_fine(self, due_date):
        # default fine logic: $1 per day overdue
        overdue = datetime.now() - due_date
        return max(overdue.days, 0) * 1
    
    def max_books(self):
        return 3  # default max books per member
    
    def __str__(self):
        return f"Member: {self.name} | Borrowed: {len(self.borrowed_books)} books"
```

```python
# Silver Member
class SilverMember(Member):
    def max_books(self):
        return 5  # higher limit
    def calculate_fine(self, due_date):
        overdue = datetime.now() - due_date
        return max(overdue.days, 0) * 1  # same fine
# Gold Member → inherits SilverMember (Multilevel)
class GoldMember(SilverMember):
    def max_books(self):
        return 10  # highest limit
    def calculate_fine(self, due_date):
        overdue = datetime.now() - due_date
        return max(overdue.days, 0) * 0.5  # half fine for gold members
```

```python
class Library:
    def __init__(self):
        self.books = []
        self.members = []
    def add_book(self, book):
        self.books.append(book)
    def add_member(self, member):
        self.members.append(member)
    def list_books(self):
        for b in self.books:
            print(b)
    def list_members(self):
        for m in self.members:
            print(m)
```

Create books

```python
b1 = Book("Python 101", "Author A", 3)
b2 = Book("Data Science Handbook", "Author B", 2)
b3 = Book("AI Basics", "Author C", 1)
```

Create Members

```python
m1 = Member("Nisha", 101)
m2 = SilverMember("Ali", 102)
m3 = GoldMember("Sara", 103)
```

Create Library & Add Books/Members

```python
library = Library()
library.add_book(b1)
library.add_book(b2)
library.add_book(b3)

library.add_member(m1)
library.add_member(m2)
library.add_member(m3)
```

Borrow & Return Books

```python
# Normal Member borrows
m1.borrow_book(b1)  # 14 days due
m1.borrow_book(b2)
# Silver Member borrows
m2.borrow_book(b3, days=7)
# Gold Member borrows
m3.borrow_book(b2)
```

Return Books

```python
# simulate overdue by manipulating due date
m1.borrowed_books[b1] = datetime.now() - timedelta(days=3)
m1.return_book(b1)  # Should show fine
m2.return_book(b3)  # On time
m3.return_book(b2)  # On time
```

```python
print("All Books in Library:")
library.list_books()
print("\nAll Members in Library:")
library.list_members()
```
