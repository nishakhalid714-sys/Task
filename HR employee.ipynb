```python
employees = [
    {"name": "Alice", "score": 85, "experience": 5},
    {"name": "Zob", "score": 195, "experience": 3},
    {"name": "Narlie", "score": 75, "experience": 7},
    {"name": "Eavid", "score": 5, "experience": 5},
]
```

 Sort by score descending, experience descending, name ascending

```python
ranked = sorted(employees, key=lambda x: (-x['score'], -x['experience'], x['name']))
```

Print ranked employees

```python
print("Ranked Employees:")
for i, emp in enumerate(ranked, 1):
    print(f"{i}. {emp['name']} - Score: {emp['score']}, Exp: {emp['experience']} years")
```

```python
 ##Quick search by name
def search_employee(name):
    result = [e for e in ranked if e['name'].lower() == name.lower()]
    return result[0] if result else None
```

```python
##for example
print("\nSearch Result:", search_employee("Eavid"))
```
