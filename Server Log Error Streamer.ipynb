simulate log lines

```python
import random
log_lines = [
    "INFO: User login successful",
    "ERROR: Disk full",
    "WARNING: CPU usage high",
    "ERROR: Out of memory",
    "INFO: Backup completed",
    "ERROR: Disk full",
    "ERROR: Network timeout",
    "INFO: Task scheduled",
    "ERROR: Out of memory"
] * 100
```

Generator Function to Stream Critical Errors

```python
def stream_critical_errors(log_iterable):
    """
    Generator that yields only ERROR log lines
    """
    for line in log_iterable:
        if line.startswith("ERROR"):
            yield line
```

Group by Error Type and Running Count

```python
from collections import defaultdict

error_counts = defaultdict(int)

for error_line in stream_critical_errors(log_lines):
    # Extract error type after 'ERROR: '
    error_type = error_line.split("ERROR: ")[1]
    error_counts[error_type] += 1
    # Live running output
    print(f"{error_type} - Count so far: {error_counts[error_type]}")
```

```python
print("\nFinal Error Summary:")
for err, count in error_counts.items():
    print(f"{err}: {count}")
    
```
